# ***value imputation***


missing data imputation used stastics:mean median mode

```python
from sklearn.impute import SimpleImputer
```

```python
print(f'no.of null value before imputing:{df.age.isnull().sum()}')
```

```python
#missing values
imp=SimpleImputer(strategy='mean')
df['age']=imp.fit_transform(df[['age']])
print(f'no.of null value after imputing:{df.age.isnull().sum()}')
```

```python
def get_param(df):
  parameters={}
  for i in df.columns[df.isnull().any()]:
    if df[i].dtype=='float64' or df[i].dtype=='int64' or df[i].dtype=='int32':
      strategy='mean'
    else:
      strategy='most_frequent'
    missing_val=df[i][df[i].isnull()].values[0]
    parameters[i]={'missing_values':missing_val,'strategy':strategy}
  return parameters

get_param(df)
```

```python
parameters=get_param(df)
```

```python
for col, param in parameters.items():
  missing_values = param['missing_values']
  strategy = param['strategy']
  imp = SimpleImputer(missing_values=missing_values, strategy=strategy)
  df[col] = imp.fit_transform(df[[col]]).squeeze()
```
